```python
def extract_sentences(text):
    sents = text.split(".")
    return [i.strip() for i in sents if i.strip() != '']
```

```python
df = pd.read_csv('sentiment_data.csv')
```

```python
df['sentences'] = df.text.map(extract_sentences)
```
